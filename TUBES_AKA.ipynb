##PEMBUATAN DATA DUMMY

```python
import random
from dataclasses import dataclass

# Definisi struct-like object
@dataclass
class Person:
    name: str
    favorite_song: str
    favorite_artist: str
    favorite_genres: list

# Daftar nama, lagu, artis, dan genre untuk randomisasi
names = ["Ali", "Budi", "Citra", "Dewi", "Eka", "Fajar", "Gita", "Hendra", "Intan", "Joko"]
songs = ["Shape of You", "Bohemian Rhapsody", "Blinding Lights", "Someone Like You",
         "Imagine", "Hotel California", "Perfect", "Let It Be", "Smells Like Teen Spirit", "Halo"]
artists = ["Ed Sheeran", "Queen", "The Weeknd", "Adele", "John Lennon",
           "Eagles", "Beyoncé", "Nirvana", "The Beatles", "Taylor Swift"]
genres = ["Pop", "Rock", "Jazz", "Classical", "Hip-Hop", "Electronic", "R&B", "Country", "Metal", "Blues"]

# Membuat 300 nama dengan atribut acak
people = []
for _ in range(300):
    person = Person(
        name=random.choice(names) + " " + random.choice(["Putra", "Putri", "Permana", "Wijaya"]),
        favorite_song=random.choice(songs),
        favorite_artist=random.choice(artists),
        favorite_genres=random.sample(genres, random.randint(2, 4))  # Memilih 2-4 genre acak
    )
    people.append(person)

# Menampilkan beberapa data sebagai contoh
for person in people[:10]:  # Tampilkan 10 data pertama
    print(f"Name: {person.name}, Song: {person.favorite_song}, Artist: {person.favorite_artist}, Genres: {', '.join(person.favorite_genres)}")

# Optional: Menyimpan ke CSV
import pandas as pd

df = pd.DataFrame([vars(p) for p in people])  # Convert dataclass ke dict untuk DataFrame
df.to_csv("people_music_preferences_with_genres.csv", index=False)
print("Data saved to people_music_preferences_with_genres.csv")
```

```python
import random
import pandas as pd

# Daftar unik lagu, genre, dan artis
songs = [
    "Shape of You", "Bohemian Rhapsody", "Blinding Lights", "Someone Like You",
    "Imagine", "Hotel California", "Perfect", "Let It Be", "Halo", "Thinking Out Loud",
    "Smells Like Teen Spirit", "Rolling in the Deep", "Stairway to Heaven",
    "Thriller", "All of Me", "Take Me to Church", "Old Town Road",
    "Bad Guy", "Shallow", "Rockstar"
]

genres = [
    "Pop", "Rock", "Jazz", "Classical", "Hip-Hop", "Electronic", "R&B",
    "Country", "Metal", "Blues", "Alternative", "Folk", "Reggae",
    "Punk", "Dance", "Soul", "Disco", "Indie", "Funk", "Trap"
]

artists = [
    "Ed Sheeran", "Queen", "The Weeknd", "Adele", "John Lennon", "Eagles",
    "Taylor Swift", "The Beatles", "Beyoncé", "Michael Jackson", "Nirvana",
    "Lady Gaga", "Led Zeppelin", "Bruno Mars", "Sam Smith", "Hozier",
    "Lil Nas X", "Billie Eilish", "Post Malone", "Dua Lipa"
]

# Memastikan pasangan lagu unik dengan 1-3 genre dan artis
random.shuffle(songs)
random.shuffle(artists)
dataset = []
for i, song in enumerate(songs):
    song_genres = random.sample(genres, random.randint(1, 3))  # Pilih 1-3 genre acak
    dataset.append({
        "Song": song,
        "Genres": ", ".join(song_genres),
        "Artist": artists[i]  # Pastikan artis unik untuk setiap lagu
    })

# Konversi ke DataFrame dan menyimpan sebagai CSV
df = pd.DataFrame(dataset)
df.to_csv("unique_songs_with_multiple_genres_and_artists.csv", index=False)

# Menampilkan dataset
print(df)
```

##ALGORITMA ITERATIF

```python
# Fungsi rekomendasi dengan sistem scoring
def recommend_songs(person_name, favorite_song, favorite_genres, favorite_artist, dataset):
    recommendations = []

    # Sequential search melalui dataset
    for _, record in dataset.iterrows():
        song_name = record["Song"]
        genres = record["Genres"].split(", ")
        artist = record["Artist"]

        score = 0  # Inisialisasi skor lagu

        # Scoring berdasarkan kecocokan
        if song_name == favorite_song:  # Jika cocok dengan lagu favorit
            score += 5
        if artist == favorite_artist:  # Jika cocok dengan artis favorit
            score += 4
        genre_matches = sum(1 for genre in genres if genre in favorite_genres)  # Hitung genre yang cocok
        score += genre_matches * 3  # Tambahkan skor genre yang cocok

        # Jika skor lebih dari 0, tambahkan ke rekomendasi
        if score > 0:
            recommendations.append((song_name, score))

    # Urutkan rekomendasi berdasarkan skor tertinggi
    recommendations = sorted(recommendations, key=lambda x: x[1], reverse=True)

    # Hanya kembalikan daftar nama lagu (tanpa skor)
    return [song for song, score in recommendations]
```

```python
# Tambahkan kolom 'Recommendation' ke DataFrame pengguna
def add_recommendations(user_df, song_df):
    user_df["Recommendation"] = user_df.apply(
        lambda user: recommend_songs(
            user["name"], user["favorite_song"], user["favorite_genres"], user["favorite_artist"], song_df
        ),
        axis=1
    )
    return user_df
```

```python
dfpeople = add_recommendations(dfpeople, dfsong)
dfpeople['Recommendation']
```

```python
# Fungsi untuk mengukur performa dengan menambah pengguna secara bertahap
def measure_performance(user_df, song_df):
    user_counts = []
    execution_times = []

    # Menambahkan pengguna secara bertahap
    for num_users in range(10, len(user_df) + 1, 10):  # Mulai dari 10, bertambah 10 hingga 300
        user_subset_df = user_df.iloc[:num_users]  # Ambil subset pengguna sesuai dengan jumlah
        # Ukur waktu eksekusi
        start_time = time.time()
        add_recommendations(user_subset_df, song_df)
        execution_times.append(time.time() - start_time)
        user_counts.append(num_users)

    return user_counts, execution_times
```

```python
# Ukur performa
user_counts, execution_times = measure_performance(dfpeople, dfsong)

# Plotting hasil
plt.figure(figsize=(8, 6))
plt.plot(user_counts, execution_times, marker='o', color='b')
plt.title("Algoritma Iteratif")
plt.xlabel("Number of Users in Dataset")
plt.ylabel("Execution Time (seconds)")
plt.grid(True)
plt.show()
```

##ALGORITMA REKURSIVE

```python
# Fungsi rekomendasi rekursif dengan sistem scoring
def recommend_songs_recursive(person_name, favorite_song, favorite_genres, favorite_artist, dataset, index=0, recommendations=None):
    # Inisialisasi list rekomendasi pada pemanggilan pertama
    if recommendations is None:
        recommendations = []

    # Basis rekursi: jika indeks melebihi jumlah baris dataset
    if index >= len(dataset):
        # Urutkan rekomendasi berdasarkan skor tertinggi
        recommendations = sorted(recommendations, key=lambda x: x[1], reverse=True)
        # Kembalikan hanya daftar lagu
        return [song for song, score in recommendations]

    # Ambil data lagu pada baris saat ini
    record = dataset.iloc[index]
    song_name = record["Song"]
    genres = record["Genres"].split(", ")
    artist = record["Artist"]

    # Hitung skor kecocokan
    score = 0
    if song_name == favorite_song:  # Skor untuk judul lagu favorit
        score += 5
    if artist == favorite_artist:  # Skor untuk artis favorit
        score += 4
    genre_matches = sum(1 for genre in genres if genre in favorite_genres)  # Hitung genre yang cocok
    score += genre_matches * 3  # Tambahkan skor genre

    # Jika skor > 0, tambahkan lagu dan skor ke daftar rekomendasi
    if score > 0:
        recommendations.append((song_name, score))

    # Rekursi ke baris berikutnya
    return recommend_songs_recursive(person_name, favorite_song, favorite_genres, favorite_artist, dataset, index + 1, recommendations)
```

```python
# Tambahkan kolom 'Recommendation' ke DataFrame pengguna
def add_recommendationsrekursive(user_df, song_df):
    user_df["Recommendation"] = user_df.apply(
        lambda user: recommend_songs_recursive(
            user["name"], user["favorite_song"], user["favorite_genres"], user["favorite_artist"], song_df
        ),
        axis=1
    )
    return user_df
```

```python
# Fungsi untuk mengukur performa dengan menambah pengguna secara bertahap
def measure_performancerekursive(user_df, song_df):
    user_counts = []
    execution_times = []

    # Menambahkan pengguna secara bertahap
    for num_users in range(10, len(user_df) + 1, 10):  # Mulai dari 10, bertambah 10 hingga 300
        user_subset_df = user_df.iloc[:num_users]  # Ambil subset pengguna sesuai dengan jumlah

        # Ukur waktu eksekusi
        start_time = time.time()
        add_recommendationsrekursive(user_subset_df, song_df)
        execution_times.append(time.time() - start_time)
        user_counts.append(num_users)

    return user_counts, execution_times
```

```python
# Ukur performa
user_counts, execution_times = measure_performancerekursive(dfpeople, dfsong)

# Plotting hasil
plt.figure(figsize=(8, 6))
plt.plot(user_counts, execution_times, marker='o', color='b')
plt.title("Algoritma Rekursive")
plt.xlabel("Number of Users in Dataset")
plt.ylabel("Execution Time (seconds)")
plt.grid(True)
plt.show()
```
